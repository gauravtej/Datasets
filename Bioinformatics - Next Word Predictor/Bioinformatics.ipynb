```python
with open('/content/CDR_TrainingSet.PubTator.txt', 'r') as file:
  content = file.read()
```

```python
content
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([content])
```

```python
tokenizer.word_index
```

```python
for sentences in content.split('.'):
  print(sentences)
```

```python
for sentence in content.split('.'):
  print(tokenizer.texts_to_sequences([sentence])[0])
```

```python
input_sequence = []
for sentence in content.split('.'):
  tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1, len(tokenized_sentence)):
    input_sequence.append(tokenized_sentence[:i+1])
```

```python
input_sequence
```

```python
max([len(x) for x in input_sequence])
```

```python
max_len_sentence = max([len(x) for x in input_sequence])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_sequence = pad_sequences(input_sequence, maxlen=max_len_sentence, padding = 'pre')
```

```python
X=padded_input_sequence[:,:-1]
```

```python
y=padded_input_sequence[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
len(tokenizer.word_index)
```

```python
y=to_categorical(y, num_classes = 12538)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, LSTM
```

```python
model = Sequential()
model.add(Embedding(12538,100,input_length=415))
model.add(LSTM(100))
model.add(Dense(12538, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=50)
```

```python
import time
```

```python
text = 'Delirium during clozapine treatment'
```

```python
import numpy as np
```

```python
for i in range(10):
  #tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  #padding
  padded_token_text = pad_sequences([token_text], maxlen=415,padding='pre')
  #predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text+ ' '+ word
      print(text)
      time.sleep(2)
```
